2.5 Ewaluacja wyników detekcji obiektów pierwszoplanowych


```python

# P = (TP/(TP + FP))
# R = (TP/(TP + FN))
# F1 = (2*P*R)/(P+R)
```

```python
image_pattern = "highway/highway/input/in000001.jpg"
gt_pattern = "highway/highway/groundtruth/gt%06d.png" 

# Open image sequence as if it's a video
cap = cv2.VideoCapture(image_pattern, cv2.CAP_IMAGES)
if not cap.isOpened():
    print("Error: cannot open image sequence")
    exit()
prev_frame = None

# do ewaluacji
frame_idx = 1
tp_total = 0
fp_total = 0
fn_total = 0
tn_total = 0

while True:

    ret, frame = cap.read()
    
    if prev_frame is None: 
        IG = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # IG = IG.astype('int')
        prev_frame = IG
        continue
    
    if not ret:
        break  # No more images
    
    # Wczytaj Ground Truth dla bieżącej klatki
    gt_path = gt_pattern % frame_idx
    gt_frame = cv2.imread(gt_path, cv2.IMREAD_GRAYSCALE)
    
    IG = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # IG = IG.astype('int')
    abs_diff = cv2.absdiff(IG, prev_frame)
    prev_frame = IG

    (T , thresh) = cv2 . threshold(abs_diff, 30, 255, cv2.THRESH_BINARY)
    
    #FILTERING
    # filtered = cv2.GaussianBlur(thresh, (15, 15), 0)
    filtered = cv2.medianBlur(thresh, 5)
    #EROSION
    kernel1 = np.ones((1, 2), np.uint8)
    eroded = cv2.erode(filtered, kernel1, iterations=3) 
    #DILATATION
    kernel2 = np.ones((3, 3), np.uint8)
    img_dilation = cv2.dilate(eroded, kernel2, iterations=2)
    #CLOSING
    kernel3 = np.ones((5, 5), np.uint8)
    closing = cv2.morphologyEx(img_dilation, cv2.MORPH_CLOSE, kernel3, iterations=3)

    retval, labels, stats, centroids = cv2.connectedComponentsWithStats(closing)
    
    
    # cv2 . imshow ("Labels", np.uint8 ( labels / retval * 255) )
    # cv2.imshow("Image Sequence", img_dilation)
    
    I_VIS = labels # copy of the input image
    if (stats.shape[0] > 1) : # are there any objects
        tab = stats [1: ,4] # 4 columns without first element
        pi = np . argmax ( tab ) # finding the index of the largest item
        pi = pi + 1 # increment because we want the index in stats , not in tab
        # drawing a bbox
        cv2.rectangle(I_VIS,(stats[pi,0],stats[pi,1]),(stats[pi,0]+stats[pi,2],stats [pi,1]+ stats [pi ,3]) ,(255 ,0 ,0) ,2)
        # print information about the field and the number of the largest element
        cv2.putText(I_VIS,"%f"%stats[pi,4],(stats[pi,0],stats[pi,1]),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,0,0))
        cv2.putText ( I_VIS,"%d"%pi,(np.int32(centroids[pi,0]),np.int32(centroids[pi,1])), cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0))

    cv2.imshow("Labels",np.uint8(I_VIS/retval*255))
    
    # --- EWALUACJA ---
    if gt_frame is not None:
        # Zakładamy: GT 255 to obiekt, 0 to tło. 
        # Czasami w GT wartości > 0 to obiekty, a 0 to tło.
        valid_mask = (gt_frame == 0) | (gt_frame == 255)

        binary_gt = (gt_frame == 255)
        binary_res = (closing > 0)

        tp = np.sum((binary_res == 1) & (binary_gt == 1) & valid_mask)
        fp = np.sum((binary_res == 1) & (binary_gt == 0) & valid_mask)
        fn = np.sum((binary_res == 0) & (binary_gt == 1) & valid_mask)
        tn = np.sum((binary_res == 0) & (binary_gt == 0) & valid_mask)

        tp_total += tp
        fp_total += fp
        fn_total += fn
        tn_total += tn
        
    frame_idx += 1
    
    if cv2.waitKey(50) & 0xFF == 27:  # Esc key to exit early
        break
    
# --- OBLICZENIA KOŃCOWE ---
# Zabezpieczenie przed dzieleniem przez zero
precision = tp_total / (tp_total + fp_total) if (tp_total + fp_total) > 0 else 0
recall = tp_total / (tp_total + fn_total) if (tp_total + fn_total) > 0 else 0
f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"Wyniki ewaluacji:")
print(f"Precision (P): {precision:.4f}")
print(f"Recall (R):    {recall:.4f}")
print(f"F1-Score:      {f1:.4f}")

cap.release()
cv2.destroyAllWindows()
```

```python
def morphology_operation(frame, prev_frame, version=1, threshold=30, filter="median", erode_kernel=(2,2), erode_iter=1, dilate_kernel=(2,2), dilate_iter=1):
    IG = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # IG = IG.astype('int')
    abs_diff = cv2.absdiff(IG, prev_frame)
    (T , thresh) = cv2 . threshold(abs_diff, threshold, 255, cv2.THRESH_BINARY)
    
    #FILTERING
    if filter == "median":
        filtered = cv2.medianBlur(thresh, 5)
    else:
        filtered = cv2.GaussianBlur(thresh, (15, 15), 0)
        
    if version == 1:
        #DILATION
        kernel2 = np.ones(dilate_kernel, np.uint8)
        dilation = cv2.dilate(filtered, kernel2, iterations=dilate_iter)
        #EROSION
        kernel1 = np.ones(erode_iter, np.uint8)
        result = cv2.erode(dilation, kernel1, iterations=erode_iter) 
        
    else:
        #EROSION
        kernel1 = np.ones(erode_iter, np.uint8)
        eroded = cv2.erode(dilation, kernel1, iterations=erode_iter) 
        #DILATATION
        kernel2 = np.ones(dilate_kernel, np.uint8)
        result = cv2.dilate(filtered, kernel2, iterations=dilate_iter)
        
    return result
    
    
```
